Imbalanced Dataset 

two techniques

1. Upsampling 
2. Down sampling

```python
import numpy as np 
import pandas as pd 
```

```python
np.random.seed(123)

n_sample=1000
class_0_ratio=0.9
n_class_0=int(n_sample * class_0_ratio)
n_class_1=int(n_sample-n_class_0)
```

```python
n_class_0
```

```python
n_class_1
```

```python
class_0=pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
    
    
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1] * n_class_1
    
    
})
```

```python
df = pd.concat([class_0, class_1], ignore_index=True)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
#upsampling 
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
```

```python
upsampled=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
```

```python
df_upsampled=pd.concat([upsampled,df_majority])
```

```python
df_upsampled['target'].value_counts()
```

```python
np.random.seed(123)

n_sample=1000
class_0_ratio=0.9
n_class_0=int(n_sample * class_0_ratio)
n_class_1=int(n_sample-n_class_0)

class_0=pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
    
    
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1] * n_class_1
    
    
})
df = pd.concat([class_0, class_1], ignore_index=True)
```

```python
df['target'].value_counts()
```

```python
#downsampling 
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
downsampled=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)
```

```python
df_downsampled=pd.concat([downsampled,df_minority])
```

```python
df_downsampled['target'].value_counts()
```
